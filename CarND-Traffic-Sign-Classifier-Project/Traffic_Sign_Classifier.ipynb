````python
# Next, lets normalize the data

def normalize_data(X):
    return (X - np.mean(X)) / np.std(X)

X_train_normalized = normalize_data(X_train_gray)
X_test_normalized = normalize_data(X_test_gray)
X_valid_normalized = normalize_data(X_valid_gray)

print("X_train_gray mean:", np.mean(X_train), " ---> X_train_normalized mean:", np.mean(X_train_normalized))
print("X_test_gray mean:", np.mean(X_test), " ---> X_test_normalized mean:", np.mean(X_test_normalized))
print("X_valid_gray mean:", np.mean(X_valid), " ---> X_valid_normalized mean:", np.mean(X_valid_normalized))


X_train = X_train_normalized
X_test = X_test_normalized 
X_valid = X_valid_normalized
````

### Model Architecture

````python
### Define your architecture here.
### Feel free to use as many code cells as needed.

import tensorflow as tf
from tensorflow.contrib.layers import flatten

# Architecture based on LeNet from the tensorflow LeNet lab and
# on http://yann.lecun.com/exdb/publis/pdf/sermanet-ijcnn-11.pdf.

def ModifiedLeNet(X):    
    # Hyperparameters
    mu = 0
    sigma = 0.1
    #############################################################################################
    
    # Layer 1: Convolutional. 
    
    # Weights and bias. The shape is (5, 5, 1, 16) due to grayscale input, NOT RGB
    conv1_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 1, 6), mean = mu, stddev = sigma))
    conv1_b = tf.Variable(tf.zeros(6))
    
    # Input = 32x32x1. Output = 28x28x6.
    conv1_X = tf.nn.conv2d(X, conv1_W, strides=[1, 1, 1, 1], padding='VALID')
    conv1_X = tf.nn.bias_add(conv1_X, conv1_b)
    
    # Activation.
    conv1_X = tf.nn.relu(conv1_X)
    
    # Pooling. Input = 28x28x6. Output = 14x14x6.
    conv1_X = tf.nn.max_pool(conv1_X, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    
    #############################################################################################
    
    # Layer 2: Convolutional. 
    
    # Weights and bias
    conv2_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 6, 16), mean = mu, stddev = sigma))
    conv2_b = tf.Variable(tf.zeros(16))
    
    # Input 14x14x6. Output = 10x10x16.
    conv2_X = tf.nn.conv2d(conv1_X, conv2_W, strides=[1, 1, 1, 1], padding='VALID')
    conv2_X = tf.nn.bias_add(conv2_X, conv2_b)
                     
    # Activation.
    conv2_X = tf.nn.relu(conv2_X)

    # TODO: Pooling. Input = 10x10x16. Output = 5x5x16.
    conv2_X = tf.nn.max_pool(conv2_X, ksize=[1, 2, 2, 1], strides=[1, 2, 2, 1], padding='VALID')
    
    #############################################################################################
    
    # Layer 3: Convolutional. 
    
    # Weights and bias
    conv3_W = tf.Variable(tf.truncated_normal(shape=(5, 5, 16, 400), mean = mu, stddev = sigma))
    conv3_b = tf.Variable(tf.zeros(400))
    
    # Input = 5x5x16. Output = 1x1x400.
    conv3_X = tf.nn.conv2d(conv2_X, conv3_W, strides=[1, 1, 1, 1], padding='VALID')
    
    conv3_X = tf.nn.bias_add(conv3_X, conv3_b)
                     
    # TODO: Activation.
    conv3_X = tf.nn.relu(conv3_X)
    
    # Flatten layer 2. Input = 5x5x16. Output = 5*5*16 = 400.
    conv2_X = flatten(conv2_X)
    
    # Flatten layer 3. Input = 1x1x400. Output = 1 * 1 * 400 = 400.
    conv3_X = flatten(conv3_X)
    
    # Concat layer 2 and 3. Input = 400 + 400. Output = 800
    # We do this so that the input to the classifier is both the output from layer 2 and 3
    # The main idea here is that the classifier will be provided with different scales of receptive fields
    # Receptive field: the region in the input space that a particular CNN’s feature is looking at (i.e. be affected by). 
    conv23_X = tf.concat_v2([conv3_X, conv2_X], 1)
    
    # Dropout to help counter overfitting. 
    conv23_X = tf.nn.dropout(conv23_X, keep_prob)
    
    
    #############################################################################################
        
    # Layer 4: Fully Connected. (Classifier) 

    # Weights and bias
    conv4_W = tf.Variable(tf.truncated_normal(shape=(800, 43), mean = mu, stddev = sigma))
    conv4_b = tf.Variable(tf.zeros(43)) 
    
    # Input = 800. Output = 43.
    logits = tf.add(tf.matmul(conv23_X, conv4_W), conv4_b)

    
    return logits

````

````python
tf.reset_default_graph() 

x = tf.placeholder(tf.float32, (None, 32, 32, 1))
y = tf.placeholder(tf.int32, (None))
keep_prob = tf.placeholder(tf.float32) # probability to keep units
one_hot_y = tf.one_hot(y, 43)
````

### Train, Validate and Test the Model

A validation set can be used to assess how well the model is performing. A low accuracy on the training and validation
sets imply underfitting. A high accuracy on the training set but low accuracy on the validation set implies overfitting.

````python
### Train your model here.
### Calculate and report the accuracy on the training and validation set.
### Once a final model architecture is selected, 
### the accuracy on the test set should be calculated and reported as well.
### Feel free to use as many code cells as needed.

from sklearn.utils import shuffle

LEARNING_RATE = 0.0009
EPOCHS = 65
BATCH_SIZE = 100

train = True

logits = ModifiedLeNet(x)

cross_entropy = tf.nn.softmax_cross_entropy_with_logits(logits, one_hot_y)
loss_operation = tf.reduce_mean(cross_entropy)
optimizer = tf.train.AdamOptimizer(learning_rate = LEARNING_RATE)
training_operation = optimizer.minimize(loss_operation)

correct_prediction = tf.equal(tf.argmax(logits, 1), tf.argmax(one_hot_y, 1))
accuracy_operation = tf.reduce_mean(tf.cast(correct_prediction, tf.float32))
saver = tf.train.Saver()

# Taken from the tensorflow LeNet lab, sligthly modified to accomodate dropout
def evaluate(X_data, y_data):
    num_examples = len(X_data)
    total_accuracy = 0
    sess = tf.get_default_session()
    for offset in range(0, num_examples, BATCH_SIZE):
        batch_x, batch_y = X_data[offset:offset+BATCH_SIZE], y_data[offset:offset+BATCH_SIZE]
        accuracy = sess.run(accuracy_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 1.0})
        total_accuracy += (accuracy * len(batch_x))
    return total_accuracy / num_examples

if train == True:

    with tf.Session() as sess:
        sess.run(tf.global_variables_initializer())
        num_examples = len(X_train)

        print("Training...")
        print()
        for i in range(EPOCHS):
            X_train, y_train = shuffle(X_train, y_train)
            for offset in range(0, num_examples, BATCH_SIZE):
                end = offset + BATCH_SIZE
                batch_x, batch_y = X_train[offset:end], y_train[offset:end]
                sess.run(training_operation, feed_dict={x: batch_x, y: batch_y, keep_prob: 0.5})

            validation_accuracy = evaluate(X_valid, y_valid)
            test_accuracy = evaluate(X_test, y_test)

            print("EPOCH {} ...".format(i+1))
            print("Test Accuracy = {:.3f}".format(test_accuracy))
            print("Validation Accuracy = {:.3f}".format(validation_accuracy))
            print()

        saver.save(sess, 'lenet')
        print("Model saved")
````
